```keyring``` allows you to store your passwords without them ever being exposed to github or any file storage. 

Usage:
In the first instance, set the password by:

```keyring.set_password("system", "username", "password")```

In the future, you can just get the password by 

```keyring.get_password("system", "username")```

For valispace, the ```system``` is simply ```valispace```.
